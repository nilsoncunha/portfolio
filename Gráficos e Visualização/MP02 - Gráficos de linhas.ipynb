```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = []
y = []
y2 = []
```

```python
for num in range(-200, 201, 10):
    x.append(num)
    y.append(num)
    y2.append(2 * num)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y, c='green', ls='-') #ls -> Line Style
```

```python
plt.plot(x, y, c='green', ls='--') #ls -> Line Style
```

**Inserindo marcadores**

```python
plt.plot(x,y, c='purple', ls='-', lw=1, marker='o')
```

**Alterando o tamanho dos marcadores**

```python
plt.plot(x,y, c='blue', ls='-', lw=2, marker='o', ms=10) # ms -> Marker Size
```

*Customizando Markers*

```python
plt.plot(x, y, c='purple', ls='-', lw=2, marker='o', ms=20, fillstyle='bottom')

# Tipos de fillstyle -> full left right bottom top none
```
