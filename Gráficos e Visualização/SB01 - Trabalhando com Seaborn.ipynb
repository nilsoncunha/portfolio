```python
import seaborn as sb
import matplotlib.pyplot as plt
import random as rd
%matplotlib inline
```

***Seaborn já possui Datasets para estudar***

```python
sb.get_dataset_names()
```

```python
titanic = sb.load_dataset('titanic')
titanic.head()
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
fly = sb.load_dataset('flights')
fly.head()
```

```python
fly = fly.pivot_table(values='passengers', index='month', columns='year')
fly
```

**Tipos de Gráficos**

*Histograma*

```python
tips.head()
```

```python
#distplot
# Mostra o histograma
sb.distplot(tips['tip'],bins=10,kde=False) #kde=False -> Remove a curva de variação
```

```python
# Aumentando os bins, diminuimos o intervalo
sb.distplot(tips['tip'],kde=False,bins=20)
```

**Correlacionando Gráficos - Gráficos Lineares**

***Joinplot***

*Junta histogramas com plots e é util para ver a relação de duas variáveis*

```python
# Verificando total_bill x tips

sb.jointplot(x='total_bill', y='tip', data=tips)
```

*Verificando valor da passagem com Idade*

```python
sb.jointplot(x='fare',y='age',data=titanic)
```

*Tamanho da mesa e tips*

```python
sb.jointplot(x='size',y='tip',data=tips)
```

**Olhando os dados de maneira geral com Pairplot**

```python
sb.pairplot(tips)
```

*Segmentando os dados com HUE*

```python
sb.pairplot(tips,hue='time')
```

```python
sb.pairplot(tips, hue='smoker')
```

*Alterando a peleta de cores*

```python
# https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/
sb.pairplot(tips, hue='sex', palette='PuOr')
```

**Plotando dados categóricos**

```python
sb.countplot(y='fare',data=titanic)
```

```python
sb.countplot(x='sex',data=titanic)
```

**Boxplot**

```python
sb.boxplot(x='day',y='tip',data=tips)
```

*Separando com HUE por sexo*

```python
sb.boxplot(x='day', y='tip', data=tips, hue='sex')
```

*Violinplot - Ideal para ver concentrações*

```python
sb.violinplot(x='day', y='tip', data=tips)
```

```python
# Utilizando o Hue
sb.violinplot(x='day', y='tip', hue='sex', data=tips)
```

```python
# Refinando o gráfico
sb.violinplot(x='day', y='tip', hue='sex', split='True', data=tips)
```

**Stripplot**

```python
sb.stripplot(x='day', y='tip', data=tips)
```

```python
sb.stripplot(x='day', y='tip', data=tips, jitter=True)
```

```python
sb.swarmplot(x='day', y='tip',data=tips)
```

```python
# Como o violin é criado

sb.swarmplot(x='day',y='tip',data=tips,color='black')
sb.violinplot(x='day',y='tip',data=tips)
```
