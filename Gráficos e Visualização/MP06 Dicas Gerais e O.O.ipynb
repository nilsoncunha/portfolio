```python
import matplotlib.pyplot as plt
import random as rd
```

```python
%matplotlib inline
```

```python
dias = []

for i in range(1,366):
    dias.append(i)
    
func01 = []
func02 = []

for i in range(1,366):
    x = rd.randint(0,100)
    func01.append(x)
    
    y = rd.randint(0,100)
    func02.append(y)
    
print(dias)
print(func01)
print(func02)
```

```python
plt.plot(dias,func01)
plt.plot(dias,func02)
```

*Fazendo a dimenso por meses*

```python
meses = []

for i in range(1,31):
    meses.append(i)
    
func01 = []
func02 = []
func03 = []

for i in range(1,31):
    x = rd.randint(0,100)
    func01.append(x)
    
    y = rd.randint(0,100)
    func02.append(y)
    
    z = rd.randint(25,70)
    func03.append(z)
    
print(meses)
print(func01)
print(func02)
print(func03)
```

```python
plt.plot(meses,func01, ls=':')
plt.plot(meses,func02, ls='-.')
plt.plot(meses,func03)
```

*Alterando o estilo do ttulo*

```python
plt.text(2,2,"texto do grfico") # text(Altura_horizontal, Altura_vertical)

plt.plot(meses,func01, ls=':')
plt.plot(meses,func02, ls='-.')
plt.plot(meses,func03)
```

*Adicionando uma Box*

```python
plt.text(2,2,"Ttulo do texto", fontsize=15, color='gray', style='oblique', weight='heavy',
         bbox=dict(facecolor='yellow', alpha=.3))

plt.plot(meses,func01)
plt.plot(meses,func02)
plt.plot(meses,func03)
```

*Adicionando anotaes*

```python
# xy -> ponto que quero apontar
# xytext -> onde que o texto vai ficar
# s -> Ttulo
plt.annotate(s="Anotao", xy=(19,90), xytext=(10,100), arrowprops=dict(arrowstyle='-'))

plt.plot(meses,func01, ls=':')
plt.plot(meses,func02, ls='-.')
plt.plot(meses,func03)
```

***Grficos Orientados a Objetos***

*Criando multiplos grficos*

```python
meses = []

for i in range(1,31):
    meses.append(i)
    
func01 = []
func02 = []
func03 = []
func04 = []

for i in range(1,31):
    x = rd.randint(0,100)
    func01.append(x)
    
    y = rd.randint(0,100)
    func02.append(y)
    
    z = rd.randint(25,70)
    func03.append(z)
    
    h = rd.randint(0,100)
    func04.append(h)
    
print(meses)
print(func01)
print(func02)
print(func03)
print(func04)
```

```python
# Multigrficos - (1,3,1) Uma linha / Trs colunas / Sequncia do grfico
plt.subplot(2,2,1)
plt.plot(meses,func01, color='green')

plt.subplot(2,2,2)
plt.plot(meses,func02, color='purple')

plt.subplot(2,2,3)
plt.plot(meses,func03, color='red')

plt.subplot(2,2,4)
plt.plot(meses,func04, ls=':')
```

*Trabalhando com a orientao a objetos nos grficos*

```python
x = range(1,13)
y = [34,65,32,65,67,89,35,65,45,89,56,45]

fig = plt.figure()

axes01 = fig.add_axes([0.1,0.1,2,0.9]) # left, bottom, rigth, top
axes01.plot(x,y)
axes01.set_xlabel('Meses')
axes01.set_ylabel('Produo')
axes01.set_title('Teste')
```

```python
meses = range(1,13)
func01 = [34,65,32,65,67,89,35,65,45,89,56,45]
func02 = [34,36,32,32,60,23,30,35,26,28,24,30]

fig = plt.figure()

axes01 = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, rigth, top
axes02 = fig.add_axes([.13,.56,0.3,0.3])

axes01.plot(meses,func01)
axes02.plot(meses,func02, c='red')


axes01.set_xlabel('Meses')
axes01.set_ylabel('Produo')
axes01.set_title('Teste')
```

*Plotando dois grficos no mesmo quadro*

```python
fig, ax = plt.subplots(set())

# Plotanto o grfico 1 e o grfico 2

ax.plot(meses,func01)
ax.plot(meses,func02)
```

*Dividindo o grfico em linhas e colunas*

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5)) # nrows, ncols

ax[0].plot(meses,func01)
ax[1].plot(meses,func02)
```
