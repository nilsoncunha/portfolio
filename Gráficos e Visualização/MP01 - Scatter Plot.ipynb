```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

**Scatter Plot**

```python
a = []
b = []

plt.scatter(a,b)
```

```python
# Escalas são diferentes de pares
a = list(np.arange(1,14))
b = list(np.arange(2,27,2))

plt.scatter(a,b)
```

**Gráfico multi linhas**

```python
x = []
y = []
y2 = []
y3 = []

for i in range(-50,100,5):
    x.append(i)
    y.append(i)
    y2.append(i * 2)
    y3.append(i + 50)
```

```python
plt.scatter(x, y, s=100, c='gray', marker='o')
plt.scatter(x, y2, s=100, c='purple', marker='+')
plt.scatter(x, y3, s=100, c='blue', marker='^')
```
