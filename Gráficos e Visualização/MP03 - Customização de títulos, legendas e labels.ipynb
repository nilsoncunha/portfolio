```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = []
y = []
y2 = []
```

```python
for num in range(-200,201,10):
    x.append(num)
    y.append(num)
    y2.append(num * 2)
```

```python
plt.plot(x, y, c='blue', ls=':', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x, y2, c='green')
```

*Adicionando descrição aos eixos*

```python
plt.plot(x,y, c='blue', ls=':', lw=2, marker='o', ms=10, fillstyle='top', mfc='blue')
plt.plot(x,y2)

# Primeira descrição
plt.xlabel('Eixo X')
```

*Formatando a fonte descriiva dos eixos*

```python
plt.plot(x,y, c='green', ls=':', lw=2, marker='o')
plt.plot(x,y2)

plt.xlabel('Espaço para o eixo X', color='red', fontsize=14)
```

*Alinhamento dos textos descritivos*

```python
# Espaço livre a partir do ponto médio do eixo
plt.plot(x,y, c='orange', ls=':', lw=2, marker='^')
plt.plot(x,y2)

#Alinhando a partir de 0
plt.xlabel('Espaço para eixo X', color='red', fontsize=15, horizontalalignment='right')
```

```python
# Y label
plt.plot(x,y, c='blue', ls=':', lw=2, marker='o')
plt.plot(x,y2)

plt.xlabel('Eixo X personalizado', color='green', fontsize=15, horizontalalignment='center')
plt.ylabel('Eixo Y personalizado', color='blue', fontsize=15, horizontalalignment='center')
```

*Adicionando título*

```python
plt.plot(x, y, c='green', ls='-', lw=2, marker='^')
plt.plot(x,y2, c='orange', ls=':', lw=1)

plt.xlabel('Título X', color='gray', fontsize=15)
plt.ylabel('Título Y', color='gray', fontsize=15)

#Title
plt.title('Título do gráfico', color='gray', fontsize=19)
```

```python
plt.plot(x,y, c='purple', ls=':', lw=3)
plt.plot(x,y2, c='red', ls='-.')

plt.xlabel('Título x', color='gray')
plt.ylabel('Título y', color='gray')

plt.title(r'Título $\frac{4}{3}$', color='gray', fontsize=20)
```

*Labels ou legendas*

```python
plt.plot(x,y, c='purple', ls='-.', label='Primeiro')
plt.plot(x,y2, c='orange', ls='-', lw=2, label='Segundo')

plt.xlabel('Titulo x', fontsize=20)
plt.ylabel('Titulo Y', fontsize=20, color='gray')

plt.title('Título do gráfico', fontsize=20, color='gray')

#legend
plt.legend()
```

*Posição das labels*

```python
plt.plot(x,y, c='red', ls='-.', lw=2, marker='o', ms=4, label='Primeiro')
plt.plot(x,y2, label='Segundo')

plt.xlabel("Título X", fontsize=21)
plt.ylabel('Título Y', fontsize=22)

plt.title('Título do gráfico', fontsize=22, color='gray')

# Posicionando a legenda
plt.legend(loc='lower right')
```

*Customizando os eixos (tickers)*

```python
plt.plot(x,y, label='Primeiro')
plt.plot(x,y2, label='Segundo')

plt.xlabel('Meses')
plt.ylabel('Y')

plt.title('Título')

plt.legend(loc='lower center')

# Customizando eixo X
eixoX = ['Janeiro','Abril','Julho']
plt.xticks([-200,0,200],eixoX)
```

```python
plt.plot(x,y, label='Primeiro')
plt.plot(x,y2, label='Segundo')

plt.xlabel('Meses')
plt.ylabel('Valores')

plt.title('Título')

plt.legend()

#Customizando Eixo X e Y

eixoX = ['Janeiro','Abril','Julho']
eixoY = ['Fraco','Regular','Neutro','Bom','Excelente']
plt.xticks([-200,0,200],eixoX)
plt.yticks([-400,-100,0,100,400],eixoY)
```

```python
eixoX = ['01/2019','04/2019','06/2019','07/2019','10/2019']
eixoY = ['Fraco','Regular','Neutro','Bom','Excelente']

plt.plot(x,y, c='purple', ls='-.', lw=2, label='Produto A')
plt.plot(x,y2, c='red', ls=':', lw=2, label='Produto B')

plt.title('Produtos Vendidos', fontsize=24, color='gray')
plt.xlabel('Meses', fontsize=20, color='gray')
plt.ylabel('Valores', color='gray', fontsize=20)

plt.legend()

plt.xticks([-200,-100,0,100,200],eixoX)
plt.yticks([-400,-100,0,100,400],eixoY)
```

***Rotacionando os tickers***

```python
eixoX = ['01/2019','04/2019','06/2019','07/2019','10/2019']
eixoY = ['Fraco','Regular','Neutro','Bom','Excelente']

plt.plot(x,y, c='purple', ls='-.', lw=2, label='Produto A')
plt.plot(x,y2, c='red', ls=':', lw=2, label='Produto B')

plt.title('Produtos Vendidos', fontsize=24, color='gray')
plt.xlabel('Meses', fontsize=20, color='gray')
plt.ylabel('Valores', color='gray', fontsize=20)

plt.legend()

plt.xticks([-200,-100,0,100,200],eixoX, rotation=-30)
plt.yticks([-400,-100,0,100,400],eixoY)
```
