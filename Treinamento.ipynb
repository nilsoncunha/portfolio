```python
df = pd.read_csv('drinks.csv')
df.head()
```

```python
df.rename(columns={'country':'paises'}, inplace=True)
df.head()
```

```python
colunas = ['paises','cerveja', 'semnome', 'vinho','total','populacao','continent']
df.columns = colunas
df.head()
```

```python
df.set_index('paises', inplace=True)
df.index.name = None
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.rename(columns={'index':'paises'},inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['populacao'] = df.populacao.str.replace('.','').astype(float)
```

```python
%matplotlib inline
df.groupby('continent').cerveja.mean().plot(kind='bar')
```

```python
index = df.groupby('continent').cerveja.mean()
cerveja = df.groupby('continent').cerveja.mean()
vinho = df.groupby('continent').vinho.sum()
plt.plot(index.index, cerveja.values)
plt.plot(index.index, vinho.values)
plt.title('Ttulo', fontsize=18, color='gray')
plt.xlabel('Rotulo X')
plt.xticks(rotation=30)
plt.ylabel('Rotulo Y')
plt.yticks(rotation=10)
```

```python
plt.rcParams['figure.figsize'] = (10,5)
```
