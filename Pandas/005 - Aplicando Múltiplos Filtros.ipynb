## Filtros Múltiplos

```python
import numpy as np
import pandas as pd
```

```python
file = 'movies.csv'
filmes = pd.read_csv(file)

filmes.columns = filmes.columns.str.replace(' ', '_')
filmes.columns
```

```python
# Filtrando com duração superior a 180 minutos

filmes[filmes.duration >= 180]
```

Regras de utilização de <b>múltiplos filtros</b> no pandas:
<ul>
    <li>use <b>&</b> em vez de <b>and</b></li>
    <li>use <b>|</b> em vez de <b>or</b></li>
    <li>Adicione <b>parênteses</b> em volta de cada critério específico, por ordem</li>
</ul>

### Trazer somente os filmes do gênero drama E com duração superior a 200 minutos.

```python
filmes.loc[(filmes['genre'] == 'Drama') & (filmes.duration >= 200)]
```

```python
dramas_longos = filmes[(filmes['genre'] == 'Drama') & (filmes.duration >= 200)]
dramas_longos
```

### Trazer somente os filmes do gênero drama OU com duração superior a 200 minutos.

```python
filmes[(filmes.genre=='Drama') | (filmes.duration >= 200)].head(20)
```

### Trazer os filmes dos gêneros Crime, Drama ou Action

```python
filmes[(filmes.genre=='Crime') | (filmes.genre=='Drama') | (filmes.genre=='Action')].head(20)
```

### Utilizando o método ISIN

```python
filmes[filmes.star_rating.isin([9.1,9.2,9.3,9.4,9.5])]
```

```python
filmes[filmes.genre.isin(['Action', 'Crime', 'Drama'])].head(20)
```
