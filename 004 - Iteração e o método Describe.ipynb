## Iterando Series e Dataframes

```python
import numpy as np
import pandas as pd
```

```python
file = 'movies.csv'
filmes = pd.read_csv(file)

filmes.columns = filmes.columns.str.replace(' ', '_')
filmes.columns
```

```python
# Imprimindo TODAS as linhas de uma s√©rie
for item in filmes.title:
    print(item)
```

```python
# Iterando um Dataframe com iterrows
for indice,linha in filmes.iterrows():
    print(indice, linha.title, ' - ', linha.genre, ' - ', linha.star_rating)
```
